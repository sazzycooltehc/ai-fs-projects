```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
tips = sns.load_dataset("tips")

# List of the five preset themes
themes = ["darkgrid", "whitegrid", "dark", "white", "ticks"]

# Set up the matplotlib figure
plt.figure(figsize=(15, 10))

for i, theme in enumerate(themes, 1):
    # Apply the theme
    sns.set_style(theme)
    
    # Create a subplot
    plt.subplot(2, 3, i)
    sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
    plt.title(f"Theme: {theme}")

# Final adjustment and display
plt.tight_layout()
plt.show()
```

```python
sns.set_style("white")
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.despine() # Removes the top and right borders
```

```python
# Figure-level example: Creates multiple subplots automatically based on 'time'
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="day", kind="scatter")
```

```python
# Axes-level example: Drawing on a specific, pre-defined spot
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.scatterplot(data=tips, x="total_bill", y="tip", ax=axes[0]) # Put in first slot
sns.histplot(data=tips, x="total_bill", ax=axes[1])            # Put in second slot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. SETUP: Apply a beautiful default style (Feature #3)
# We'll use 'whitegrid' for better readability with heavy data
sns.set_theme(style="whitegrid", palette="muted")

# 2. LOAD DATA: Seamlessly works with Pandas DataFrames (Feature #5)
df = sns.load_dataset("tips")

# --- FIGURE-LEVEL PLOT (Concept #1) ---
# relplot() creates a FacetGrid automatically. 
# It handles faceting (col="time") and statistical mapping (hue="day")
g = sns.relplot(
    data=df, x="total_bill", y="tip", 
    col="time", hue="day", style="sex",
    kind="scatter", height=5, aspect=1.2
)
g.fig.suptitle("Figure-Level: Automated Faceting (relplot)", y=1.05)

# --- AXES-LEVEL PLOTS (Concept #2) ---
# Here we manually create a layout with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Barplot (Feature #1 - Statistical Data Visualization)
# Notice it automatically calculates the mean and confidence interval (black lines)
sns.barplot(data=df, x="day", y="total_bill", ax=axes[0])
axes[0].set_title("Axes-Level: Statistical Bar Plot")

# Plot 2: Violin Plot (Feature #6 - Built-in Plot Types)
# Shows the distribution density of the data
sns.violinplot(data=df, x="day", y="total_bill", hue="sex", split=True, ax=axes[1])
axes[1].set_title("Axes-Level: Categorical Violin Plot")

# 3. CUSTOMIZATION: Using the built-on Matplotlib foundation (Feature #4)
plt.tight_layout()
sns.despine(left=True) # Feature #7: Refining the aesthetic

plt.show()
```
