```python
# Sample text
doc_4 = nlp("New York is a great city.")
# Merge "New York" into one token
with doc_4.retokenize() as retokenizer:
    span = doc_4[0:2] # Tokens "New" and "York"
retokenizer.merge(span)
# Print tokens after merging
print("After merging:")
for token in doc_4:
    print(token.text)
```

```python
from spacy.tokens import Doc
# Let's say we want to split "don't" into "do" and "n't"
words = ["I", "do", "n't", "like", "apples"]
spaces = [True, False, True, True, False] # True = space follows the token
# Create a new Doc from words
new_doc = Doc(nlp.vocab, words=words, spaces=spaces)
print("After splitting:")
for token in new_doc:
    print(token.text)
```

```python
text = "Barack Obama visited New York last week."
doc_5 = nlp(text)
with doc_5.retokenize() as retokenizer:
    for ent in doc_5.ents:
        retokenizer.merge(ent)
    print("After merging named entities:")
    for token in doc_5:
        print(token.text, "->", token.ent_type_)
```

```python
doc_7 = nlp('I want to fly from Mumbai to Delhi.')
for token in doc_7:
    if token.dep_ == 'ROOT': intent = token.text
    if token.dep_ == 'pobj' and token.head.text == 'from': departure = token.text
    if token.dep_ == 'pobj' and token.head.text == 'to': destination = token.text
print("Intent: "+intent, "Departure: "+departure, "Destination: "+destination)
```

```python
text = '''
Nuslias are used for the production of high fructose syrup and 
pharma industries.
Aspergillus in shake flask fermentation.
First report on properties of Anega 442.
'''
```

```python
import re
text = text.lower()
text = re.sub(r'\W+',' ', text)
text = re.sub(r'\d+', '', text)
```

```python
text
```

```python
from collections import Counter
def remove_rare_words(text, threshold=2):
    words = nltk.word_tokenize(text)
    freq = Counter(words)
    filtered = [w for w in words if freq[w] >= threshold]
    return ' '.join(filtered)
text = remove_rare_words(text)
```

```python
text
```

```python
nltk.download('stopwords')
```

```python
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text)
filtered_tokens = [t for t in tokens if t not in stop_words]
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in 
filtered_tokens]
```

```python
#Case Study 11 - Airlines

import pandas as pd
import spacy
from collections import Counter
# Load spaCy model
nlp_4 = spacy.load("en_core_web_md")
# Sample airline reviews (replace with actual dataset
reviews = [
    "I flew from Toronto to New York on Air Canada.",
    "Delta Airlines provided great service.",
    "Lufthansa was ayed at Frankfurt airport."
]
# Extract entities
entity_counter = Counter()
for review in reviews:
    doc_cs_11 = nlp_4(review)
    for ent in doc_cs_11.ents:
        entity_counter [ent.text] += 1
# Display top 10 entities
print("Top 10 Entities:")
for entity, freq in entity_counter.most_common (10):
    print (f"{entity}: {freq}")
```

```python
from fastapi import FastAPI
from transformers import pipeline

app = FastAPI()

# Explicitly specify the model
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english",
    revision="714eb0f"  # optional, locks to a specific version
)

@app.get("/predict")
def predict(text: str):
    result = classifier(text)
    return {"result": result}
```
